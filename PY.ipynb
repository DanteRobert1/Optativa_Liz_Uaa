```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import numpy as np
import plotly.express as px
import plotly.graph_objects as go

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
df = pd.read_csv ('https://raw.githubusercontent.com/DanteRobert1/Optativa_Liz_Uaa/refs/heads/main/vgsales.csv', sep = ',' )
```

```python
print(df.head())
```

```python
df['Regional_Sales'] = df['NA_Sales'] + df['EU_Sales'] + df['JP_Sales']

# Crear un gráfico de dispersión
fig = px.scatter(
    df,
    x='Regional_Sales',  # Ventas combinadas de las regiones
    y='Global_Sales',    # Ventas globales
    color='Genre',       # Género
    size='Regional_Sales',  # Tamaño según la suma de las ventas regionales
    hover_name='Name',
    title='Dispersión de ventas de videojuegos con las Regiones Combinadas'
)
fig.show()
```
