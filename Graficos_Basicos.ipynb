```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Datos de Ejemplo

x = np.linspace(0, 10, 100)
y = np.sin(x)

#Grafico de Lineas
plt.plot(x, y)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Grafico de Lineas')
plt.show()
```

```python
#Grafico de Lineas
plt.figure(figsize=(10, 4))
plt.plot(x, y)
plt.xlabel('Grafico de Lineas')
plt.ylabel('x')
plt.title('sin(x)')
plt.show()
```

Barras

```python
#Grafico de Barras
categorias = ['A', 'B', 'C', 'D']
valores = [10, 15, 8, 12]

plt.figure(figsize=(8, 4))
plt.bar(categorias, valores)
plt.xlabel('Categorias')
plt.ylabel('Valores')
plt.title('Grafico de Barras')
plt.show()
```

Dispersion

```python
#Grafico e dispersion
x = np.random.rand(50)
y = np.random.rand(50)

plt.figure(figsize=(8, 4))
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Grafico de dispersion')
plt.show()
```

Historgama

```python
datos = np.random.randn(1000)

plt.figure(figsize=(8, 4))
plt.hist(datos, bins=30)
plt.xlabel('valores')
plt.ylabel('Frecuencia')
plt.title('Histograma')
plt.show
```
